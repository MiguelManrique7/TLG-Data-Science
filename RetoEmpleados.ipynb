Se cargan los datos:

```python
import pandas as pd
EmpleadosAttrition =  pd.read_csv('empleadosRETO.csv')
```

Se revisan la estadísticas sumarias para conocer la naturaleza de los datos:

```python
EmpleadosAttrition.describe()
```

Se eliminan algunas variables:

```python
EmpleadosAttrition =  EmpleadosAttrition.drop(columns = ['EmployeeCount', 'EmployeeNumber','Over18','StandardHours'])
```

```python
EmpleadosAttrition['Year'] = EmpleadosAttrition['HiringDate'].str.split(pat='/').str[2].astype(int)
```

```python
EmpleadosAttrition['YearsAtCompany'] = 2018 - EmpleadosAttrition['Year']
```

```python
EmpleadosAttrition.rename(columns = {'DistanceFromHome':'DistanceFromHome_km'}, inplace = True)
```

```python
EmpleadosAttrition['DistanceFromHome'] = EmpleadosAttrition['DistanceFromHome_km'].str[0:-2].astype(int)
```

```python
EmpleadosAttrition['DistanceFromHome'] = EmpleadosAttrition['DistanceFromHome_km'].str.split(pat=' ').str[0].astype(int)
```

```python
EmpleadosAttrition = EmpleadosAttrition.drop(columns = ['Year', 'HiringDate', 'DistanceFromHome_km' ])
```

```python
SueldoPromedioDepto = pd.DataFrame(EmpleadosAttrition.groupby(['Department'])['MonthlyIncome'].mean())
SueldoPromedioDepto.rename(columns = {'MonthlyIncome':'SueldoPromedio'}, inplace = True)
SueldoPromedioDepto
```

```python
EmpleadosAttrition['MonthlyIncome'] = (EmpleadosAttrition['MonthlyIncome'] - EmpleadosAttrition['MonthlyIncome'].min()) / (EmpleadosAttrition['MonthlyIncome'].max() - EmpleadosAttrition['MonthlyIncome'].min())
```

Se genera función para pasar las variables categoricas a númericas

```python
def catonum(df, col) :
    y = df[col].astype('category')
    map =  y.cat.categories.tolist()
    replace_map_comp = {col : {k: v for k,v in zip(map,list(range(0,len(map))))}}
    df.replace(replace_map_comp, inplace=True)
    return replace_map_comp
```

```python
categoricas = ['BusinessTravel','Department','EducationField','Gender','JobRole','MaritalStatus']
```

```python
for x in categoricas :
    print(x + ' corr to Attrition '  + str(EmpleadosAttrition['Attrition'].corr(EmpleadosAttrition[x])))
```

En consecuencia, bajo el criterio de mantener aquellas variables con una correlación mayor o igual a 0.1, solamente se mantiene MaritalStatus.

```python
catonum(EmpleadosAttrition,'OverTime')
EmpleadosAttritionFinal =  EmpleadosAttrition.drop(columns = categoricas[0:5] )
EmpleadosAttritionFinal = EmpleadosAttritionFinal.dropna()
```

```python
from sklearn.decomposition import PCA
pca = PCA()
pca.fit(EmpleadosAttritionFinal)
EmpleadosAttritionPCA = pca.transform(EmpleadosAttritionFinal)
```

```python
vnam = pd.Series(list(EmpleadosAttritionFinal.columns))
pvar = pd.DataFrame(pca.explained_variance_ratio_)

pvar = pd.concat([pvar, vnam],axis = 1)
pvar
```

Los primeros 3 componentes alcanzan a explicar el 80 por ciento.

```python
pvar.iloc[0:3,0].sum()
```

```python
EmpleadosAttritionFinal =  EmpleadosAttritionFinal.assign(C0 = EmpleadosAttritionPCA[:,0] )
EmpleadosAttritionFinal =  EmpleadosAttritionFinal.assign(C1 = EmpleadosAttritionPCA[:,1] )
EmpleadosAttritionFinal =  EmpleadosAttritionFinal.assign(C2 = EmpleadosAttritionPCA[:,2] )
```
